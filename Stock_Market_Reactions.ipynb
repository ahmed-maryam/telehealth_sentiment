```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
%matplotlib inline
```

```python
# Create start and end variables
start= datetime.datetime(2020,1,1)
end = datetime.datetime(2021,10,22)
```

# Gather Stock Information

```python
# UnitedHealth Group
UNH = data.DataReader('UNH','yahoo',start,end)
# American Well
AMWL = data.DataReader('AMWL','yahoo',start,end)
# Doximity
DOCS = data.DataReader('DOCS','yahoo', start,end)
# Cerner
CERN = data.DataReader('CERN','yahoo', start,end)
# Teladoc
TDOC = data.DataReader('TDOC','yahoo', start,end)
# UpHealth
UPH = data.DataReader('UPH','yahoo', start,end)
#1Life Healthcare
ONEM = data.DataReader('ONEM','yahoo', start,end)
```

```python
tickers = ['UNH','AMWL','DOCS','CERN','TDOC','UPH','ONEM'] 
```

```python
consumer_stocks = pd.concat([UNH,AMWL,DOCS,CERN,TDOC,UPH,ONEM], axis = 1, keys = tickers)
```

```python
consumer_stocks.columns.names = ['Consumer Ticker','Stock Info']
```

```python
consumer_stocks.head()
```

```python
consumer_stocks.to_csv('Consumer_Stock.csv')
```
